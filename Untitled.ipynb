```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('games.csv')
df.dtypes
```

```python
'''We don't need usernames or game ids as it is unimportant to the outcome. Including the moves will also
ruin the outcome as we would then have the answer to who won. We are only interested in the opening name which
includes the first 2-6 moves.'''
df.drop(['id','white_id', 'black_id', 'moves', 'opening_name', 'opening_ply'], axis = 1, inplace = True)
df
```

```python
#print(df.created_at[0], df.last_move_at[0])
print(df.info)
df.describe()
df.dtypes
```

```python
df['length'] = abs(df['last_move_at'] - df['created_at'])
#df.drop(['created_at', 'last_move_at'], axis = 1, inplace = True)
```

```python
df.groupby('winner').nunique()
```

```python
x = len(df['winner'])

def winner_rating_diff():
    for i in range(x):
        if df.winner[i] == 'white':
            df['wrd'] = df['white_rating'] - df['black_rating']
        elif df.winner[i] == 'black':
            df['wrd'] = df['black_rating'] - df['white_rating']
        else: 
            df['wrd'] = 'Null'
        
winner_rating_diff()
```

```python
df[['game_timemins', 'added_move_timesec']] = df.increment_code.apply(lambda x: pd.Series(str(x).split("+")))
df.drop(['increment_code', 'created_at', 'last_move_at'], axis = 1, inplace = True)
df
```

```python
df = df[(df.white_rating != 1500) & (df.black_rating != 1500)]
#to have a mate, you have to have at least 5 moves, these games would have been resigned or drawn for no reason
df = df[(df.turns) > 6]
print(df)
df.dtypes
```

```python
eco = pd.read_excel('eco_list.xlsx')
eco
eco_df = eco[['Layman_name', 'Eco_name']]
eco_df
```

```python
print(len(eco_df.groupby('Layman_name')))
eco_df.groupby('Layman_name')['Layman_name'].count()
```

```python
eco_array = eco_df['Eco_name'].to_numpy()
name_array = eco_df['Layman_name'].to_numpy()

D = {}
for i in range(len(name_array)):
    D[name_array[i]] = []
    D[name_array[i]].append(eco_array[i])
print(D)
```

```python
D = {}
eco_array = eco_df['Eco_name'].to_numpy()
name_array = eco_df['Layman_name'].to_numpy()

for i in range(len(name_array)):
    if name_array[i] in D:
        D[name_array[i]].append(eco_array[i])
    else:
        D[name_array[i]] = []
        D[name_array[i]].append(eco_array[i])
#print(D)

C = {}
for i in range(len(eco_array)):
    if eco_array[i] in C:
        C[eco_array[i]].append(name_array[i])
    else:
        C[eco_array[i]] = []
        C[eco_array[i]].append(name_array[i])
print(C)
```

```python
col = df['opening_eco'].replace(C)
df['short_name'] = col
df
```

```python
print(len(df.groupby('short_name')))
df.groupby('short_name')['short_name'].count() / len(df) * 100
counts = df.groupby('short_name')['short_name'].count()
pct = counts.to_numpy() / len(df) * 100
print(counts.to_numpy(), pct)
```

```python
inputs = df.drop('winner', axis = 'columns')

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

target = df.drop(['rated', 'turns', 'victory_status', 'white_rating', 'black_rating',
                  'opening_eco', 'length', 'wrd', 'game_timemins', 'added_move_timesec', 'short_name'], axis = 'columns')
target_le = target
target_le.winner = le.fit_transform(target_le.winner)
target_le

#black = 0, white = 2, draw = 1
```

```python
inputs
```

```python
dummies_1 = pd.get_dummies(df.victory_status)
dummies_1 = dummies_1.drop('outoftime', axis = 'columns')
dummies_2 = pd.get_dummies(df.short_name)
dummies_2 = dummies_2.drop('Vienna', axis = 'columns')

dummies_concat = pd.concat([dummies_1, dummies_2], axis=1)

final_dataset = pd.concat([inputs, dummies_concat], axis=1)
final_dataset = final_dataset.drop(['rated', 'victory_status', 'opening_eco', 'short_name'], axis = 'columns')
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
y = target_le
X = final_dataset

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state = 42)
dt_model = tree.DecisionTreeClassifier()

dt_model.fit(X_train, y_train)
y_test_dt = dt_model.predict(X_test)

from sklearn.metrics import accuracy_score
acc_dt = accuracy_score(y_test_dt, y_test)
acc_dt
```

```python
import seaborn as sns

X_turns = X["turns"]
Q1 = X_turns.quantile(0.25)
Q3 = X_turns.quantile(0.75)
IQR = Q3-Q1
alt_sinir = Q1 - 1.5*IQR
ust_sinir = Q3 + 1.5*IQR
tum_aykiriliklar = (X_turns < alt_sinir) | (X_turns > ust_sinir)
X_turns[tum_aykiriliklar] = X_turns.mean()
```

```python
def normalize(val):
    val = (val-np.min(val))/(np.max(val)-np.min(val))
    return val
X["turns"] = normalize(X["turns"])
X.head()
```

```python
y = target_le
X = final_dataset

for i in range(100):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state = 42)
    dt_model = tree.DecisionTreeClassifier()
    dt_model.fit(X_train, y_train)
    y_test_dt = dt_model.predict(X_test)
    acc_dt = accuracy_score(y_test, y_test_dt)
    print(acc_dt)
```
